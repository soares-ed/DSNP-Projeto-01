# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada por recém-descoberto coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

<center><img alt="Colaboratory logo" width="50%" src="https://raw.githubusercontent.com/carlosfab/dsnp2/master/img/covid19.jpg"></center>

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

### Dicionário de Variáveis
*  `iso_code` - Código de identificação do oaís

* `new_tests_per_thousand	` - Novos testes realizados por mil pessoas

* `new_tests_smoothed_per_thousand` - Novos testes realizados nos últimos 7 dias por mil pessoas

* `tests_units` - Unidades utilizadas pelo local para relatar seus dados de teste.
* `total_vaccinations` - Total de doses administradas da vacina contra a COVID-19
* `people_vaccinated` - Total de pessoas que receberam ao menos uma dose da vacina contra a COVID-19
* `people_fully_vaccinated` - Total de pessoas que estão completamente vacinadas (duas doses ou não) contra a COVID-19
* `new_vaccinations` - Novas doses de vacinas administradas contra a COVID-19 (apenas calculado para dias consecutivos)
* `new_vaccinations_smoothed` - Novas doses de vacinas administradas contra a COVID-19 (apenas calculado para dias consecutivos) nos últimos 7 dias
* `total_vaccinations_per_hundred` - Novas doses de vacinas administradas contra a COVID-19 (apenas calculado para dias consecutivos) por cem pessoas
* `people_vaccinated_per_hundred` - Total de pessoas que receberam ao menos uma dose da vacina contra a COVID-19 a cada cem pessoas
* `people_fully_vaccinated_per_hundred` - Total de pessoas que estão completamente vacinadas (duas doses ou não) contra a COVID-19 a cada cem pessoas
* `new_vaccinations_smoothed_per_million` - Novas doses de vacinas administradas contra a COVID-19 (apenas calculado para dias consecutivos) por 1 milhão de pessoas, nos últimos 7 dias
* `stringency_index` - Índice de Restrição de Resposta Governamental: medida composta baseada em 9 indicadores de resposta incluindo encerramentos de escolas, encerramentos de locais de trabalho e proibições de viagens, redimensionada para um valor de 0 a 100 (100 = resposta mais estrita)

* `gdp_per_capita` - Produto Interno Bruto (PIB) per capita

### Países com mais casos de covid-19 até a data mais recente
Neste momento vamos tratar os dados e descobrir os países com mais casos até a data mais recente.
Devemos remover as tuplas que contenham "OWID" dentro a variável `iso_code`. Antes de tratar os dados, será criado uma cópia do *dataset*


```
# cópia do dataframe
df_covid19 = df.copy()

# apagar tuplas em iso_code que contenha "OWID"
df_covid19.drop(df_covid19.index[df_covid19.iso_code.str.contains("OWID")].values, inplace = True)

# lista dos 5 páises com mais casos
df_covid19.loc[df.date == '2023-07-05'].sort_values(by="total_cases",ascending=False).dropna(subset=['continent'], axis=0)[['location', 'total_cases']].head()
```

### Gráfico com os 5 países com mais casos confirmado até a data atual

```
#criando o dataset do gráfico
top_5_total_cases = df_covid19.loc[df.date == '2023-07-05', ['location', 'total_cases']].sort_values(by="total_cases", ascending=False)[:5]

# atribuindo valores
fig, ax = plt.subplots(figsize=(6,4))
ax.barh('location', 'total_cases',color=['blue', 'grey', 'grey', 'grey', 'grey'], data=top_5_total_cases)

#titulos e subtitulos
plt.title('5 Países com mais casos de Covid-19')


#customização
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["left"].set_visible(False)

#plotando o gráfico
plt.tight_layout()
```

### Países com mais mortes confirmadas


```
# lista dos 5 páises com mais mortes confirmadas
df_covid19.loc[df.date == '2023-07-05'].sort_values(by="total_deaths",ascending=False).dropna(subset=['continent'], axis=0)[['location', 'total_deaths']].head()
```

```

#criando o dataset do gráfico
top_5_total_deaths = df_covid19.loc[df.date == '2023-07-05', ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[:5]

# atribuindo valores
fig, ax = plt.subplots(figsize=(6,4))
ax.barh('location', 'total_deaths',color=['red', 'grey', 'grey', 'grey', 'grey'], data=top_5_total_deaths)

#titulos e subtitulos
plt.title('5 Países com mais morte por Covid-19')


#customização
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["left"].set_visible(False)


#plotando o gráfico
plt.tight_layout()
```

### Como o vírus está progredindo?
Uma das maiores preocupações das organizações de saúde é a progressão do número de casos e mortes no mundo.

Uma das melhores formas de visualizarmos a evolução ao longo do tempo, é o gráfico de linhas. Por isso, vamos plotar um gráfico de linha para demonstrar e evolução de mortes no mundo.

```
# criando o dataset do gráfico
df_world = df.loc[df.location == "World", "total_deaths",].reset_index(drop=True)

# valores
fig, ax = plt.subplots(figsize=(10,5))
df_world.plot(x="date", y="total_death",kind='line', color="red")

# customização
ax.set_yscale('log')
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

#titulos e subtitulos
plt.title('Progressão do Covid-19 pelo mundo')
ax.set_xlabel("Dias")
ax.set_ylabel("Mortes")

# plotar
plt.tight_layout()
```

## Análise Exploratória para o Brasil

* Criar uma cópia do DataFrame apenas com o Brasil
* Identificar quando foi registrada a primeira morte no Brasil
* Identificar quantos dias demorou para registrarmos a primeira morte, considerando o primeiro caso de COVID-19 registrado.
* Plotar gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais.

```
# criar um DataFrame apenas com o Brasil
df_brasil = df.loc[df.location == "Brazil"].copy()
df_brasil.head()
```

```
#data do primeiro caso de Covid-19 no Brasil
df_brasil.loc[df_brasil['new_cases'] != 0.0].head(1)
```

```
#primeira morte
df_brasil.loc[df_brasil['new_deaths'] != 0.0].head(1)
```

- O primeiro caso de Covid-19 no Brasil foi registrado no dia **27/02/2020**
- A primreira morte por Covid-19 foi registrada no dia **19/03/2020**
- Demorou **21** dias para ser registrada a primeira morte considerando o primeiro caso de Covid-19 mo Brail


```
# Vamos analisar a evolução da pandemia no Brasil pelo número de casos

#criar um Dataframe para os casos totais no Brasil
df_brasil3 = df_brasil.loc[df_brasil.location == 'Brazil', 'total_cases',].reset_index(drop=True)

# valores
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(20,7))
df_brasil3.plot(x="date", y="total_cases",kind='line',ax=ax0, color="blue", lw=3)
df_brasil3.plot(x="date", y="total_cases",kind='line',ax=ax1, color="blue", lw=3)

# customização
ax0.set_yscale('linear')
ax0.spines["right"].set_visible(False)
ax0.spines["top"].set_visible(False)
ax0.spines['bottom'].set_visible(False)
ax0.set_xticks(ticks=[], minor=False)


ax1.set_yscale('log')
ax1.spines["right"].set_visible(False)
ax1.spines["top"].set_visible(False)
ax1.set_xlabel('Dias',fontsize=20, color='gray', fontweight='bold')

#text
ax0.set_title('Evolução dos casos de Covid-19 no Brasil',fontsize=20, color='gray', fontweight='bold')
ax0.text(0.98, 0.75,'Linear',transform=ax0.transAxes, fontsize=16, color='gray', fontweight='bold')
ax1.text(0.98, 0.80,'Logarítmica',transform=ax1.transAxes, fontsize=16, color='gray', fontweight='bold')

#plotar
plt.tight_layout()
plt.show()
```

```
df_brasil2 = df_brasil.loc[df_brasil.location == 'Brazil', 'total_deaths',].reset_index(drop=True)


# valores
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(20,7))
df_brasil2.plot(x="date", y="total_death",kind='line',ax=ax0, color="blue", lw=3)
df_brasil2.plot(x="date", y="total_death",kind='line',ax=ax1, color="blue", lw=3)

# customização
ax0.set_yscale('linear')
ax0.spines["right"].set_visible(False)
ax0.spines["top"].set_visible(False)
ax0.spines['bottom'].set_visible(False)
ax0.set_xticks(ticks=[], minor=False)


ax1.set_yscale('log')
ax1.spines["right"].set_visible(False)
ax1.spines["top"].set_visible(False)
ax1.set_xlabel('Dias',fontsize=20, color='gray', fontweight='bold')

#text
ax0.set_title('Evolução das mortes por Covid-19 no Brasil',fontsize=20, color='gray', fontweight='bold')
ax0.text(0.98, 0.75,'Linear',transform=ax0.transAxes, fontsize=16, color='gray', fontweight='bold')
ax1.text(0.98, 0.80,'Logarítmica',transform=ax1.transAxes, fontsize=16, color='gray', fontweight='bold')

#plotar
plt.tight_layout()
plt.show()
```
